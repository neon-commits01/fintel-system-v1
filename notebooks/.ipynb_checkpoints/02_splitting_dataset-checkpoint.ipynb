```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Let's load the dataset
df = pd.read_csv('../data/processed/cleaned_data.csv')
df.shape
```

```python
# Since, we need to split the data, we need a X and y. So, let's go ahead and define them
X = df['text']
y = df['sentiment']
# Splitting the dataset
X_dev, X_test, y_dev, y_test = train_test_split(X,y,test_size=0.15, random_state=42, stratify=y)
```

```python
# Convert this Series into Dataframe
dev_data = X_dev.to_frame(name='text')
dev_data['sentiment'] = y_dev
dev_data = dev_data.reset_index().rename(columns={'index': 'original_index'})
dev_data
```

```python
test_data = X_test.to_frame(name='text')
test_data['sentiment'] = y_test
test_data  = test_data.reset_index().rename(columns={'index':'original_index'})
test_data
```

```python
print("Original distribution:")
print(y.value_counts(normalize=True))

print("\nDevelopment distribution:")
print(y_dev.value_counts(normalize=True))

print("\nTest distribution:")
print(y_test.value_counts(normalize=True))

print("\nDev rows:", len(dev_data))
print("Test rows:", len(test_data))
print("Total:", len(dev_data) + len(test_data))
print("Original:", len(df))
```

```python
dev_data.to_csv('../data/processed/dev_data.csv',index=False)
test_data.to_csv('../data/processed/test_data.csv',index=False)
```
