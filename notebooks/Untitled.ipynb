```python
import joblib
```

```python
model = joblib.load("../models/tfidf_linear_svm.joblib")
```

```python
sample_text = ["Operating profit fell compared with the previous year."]
model.predict(sample_text)
```

```python
test_headlines = [
    # Should be negative
    "Company reports significant losses in Q3",
    
    # Should be positive  
    "Revenue increased to EUR 45mn from EUR 32mn last year",
    
    # Tricky — improvement from negative (model might fail)
    "Net profit was EUR 5mn compared to a loss of EUR 12mn",
    
    # Tricky — formal factual tone (model might say neutral)
    "Company has reduced 200 jobs across three units",
    
    # Should be neutral
    "The company will hold its annual general meeting next month"
]

for text in test_headlines:
    prediction = model.predict([text])
    print(f"Text: {text[:50]}...")
    print(f"Predicted: {prediction[0]}")
    print("---")
```
